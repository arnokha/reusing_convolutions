```python
def extract_avg(f):
    with open(f) as f:
        first_line = f.readline()
        return float(first_line.rsplit(" ")[-1])
```

```python
extract_avg("Alien/test_new_conv_200_out.txt")
```

```python
e_20 = "v_20_o"
e_50 = "v_50_o"
e_100 = "v_100_o"
e_200 = "v_200_o"
e_20_40 = "v_20_40_o"
e_40_80 = "v_40_80_o"
e_80_160 = "v_80_160_o"
results = {}

for game in atari_games_list:
    list_20 = []
    list_50 = []
    list_100 = []
    list_200 = []
    list_20_40 = []
    list_40_80 = []
    list_80_160 = []
    
    dir_cont = sorted(os.listdir(ex))[::-1] # standard then new
    for filename in dir_cont:
        if e_20 in filename:
            list_20.append(game + "/" + filename)
        if e_50 in filename:
            list_50.append(game + "/" + filename)
        if e_100 in filename:
            list_100.append(game + "/" + filename)
        if e_200 in filename:
            list_200.append(game + "/" + filename)
        if e_20_40 in filename:
            list_20_40.append(game + "/" + filename)
        if e_40_80 in filename:
            list_40_80.append(game + "/" + filename)
        if e_80_160 in filename:
            list_80_160.append(game + "/" + filename)
            
    ## Check for missing files
    temp = [list_20, list_50, list_100, list_200, list_20_40, list_40_80, list_80_160]
    for i,t in enumerate(temp):
        if len(t) != 2:
            print(game, i)
      
    r = {}
    r["20"] = (extract_avg(list_20[0]), extract_avg(list_20[1]))
    r["50"] = (extract_avg(list_50[0]), extract_avg(list_50[1]))
    r["100"] = (extract_avg(list_100[0]), extract_avg(list_100[1]))
    r["200"] = (extract_avg(list_200[0]), extract_avg(list_200[1]))
    r["20_40"] = (extract_avg(list_20_40[0]), extract_avg(list_20_40[1]))
    r["40_80"] = (extract_avg(list_40_80[0]), extract_avg(list_40_80[1]))
    r["80_160"] = (extract_avg(list_80_160[0]), extract_avg(list_80_160[1]))
    results[game] = r
    
    #with open(RESULTS_FILENAME) as rf:
    #    rf.write(game + ": ")
    #    rf.write( list_20)
```

```python
def plot_times(game, results):
    # create plot
    result_names = ["1 Layer,\n20 out channels", 
                    "1 Layer,\n50 out channels",
                    "1 Layer,\n100 out channels",
                    "1 Layer,\n200 out channels",
                    "2 Layers,\n20/40\nout channels",
                    "2 Layers,\n40/80\nout channels",
                    "2 Layers,\n80/160\nout channels"
                   ]
    
    standard_results = [results[game]["20"][0],
                        results[game]["50"][0],
                        results[game]["100"][0],
                        results[game]["200"][0],
                        results[game]["20_40"][0],
                        results[game]["40_80"][0],
                        results[game]["80_160"][0]]
    
    reuse_results = [results[game]["20"][1],
                     results[game]["50"][1],
                     results[game]["100"][1],
                     results[game]["200"][1],
                     results[game]["20_40"][1],
                     results[game]["40_80"][1],
                     results[game]["80_160"][1]]

    
    plt.figure(figsize=(16,8))
    index = np.arange(len(result_names))
    bar_width = 0.25
    opacity = 0.75

    rects1 = plt.bar(index, standard_results, bar_width,
    alpha=opacity,
    color='b',
    label="Standard Model")

    rects2 = plt.bar(index + bar_width, reuse_results, bar_width,
    alpha=opacity,
    color='r',
    label="Reuse Model")

    plt.xlabel("Model Architecture", fontsize=14)
    plt.ylabel("Avg. Runtime (in seconds) for 1000 game frames", fontsize=14)
    plt.title("Average Runtime for 1000 grame frames by model type (" + game + ")", fontsize=20)
    plt.xticks(index + bar_width - 0.1, result_names, fontsize=16)
    plt.legend()

    plt.tight_layout()
    plt.savefig("result_figures/" + game + "_results.png")
    plt.show()
```

```python
for game in atari_games_list:
    plot_times(game, results)
    plt.savefig(game + "_results.png")
```
